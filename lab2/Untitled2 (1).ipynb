# **Задание 1**

```python
#Импорт библиотек
import scipy.stats as sps
import numpy as np

#Выборки
X1 = sps.norm(loc=2, scale=1).rvs(size=25)
X2 = sps.norm(loc=1, scale=np.sqrt(0.5)).rvs(size=25)
print(f'Первая выборка: {X1}')
print(f'Вторая выборка: {X2}')

#Квантили
q1 = sps.norm(loc=0, scale=np.sqrt(50)/3).ppf([0.025])
q2 = sps.norm(loc=0, scale=np.sqrt(50)/3).ppf([0.975])
print(f'Первый квантиль: {q1}')
print(f'Второй квантиль: {q2}')

#Левая граница доверительного интервала
left = X1.mean() - X2.mean() - np.sqrt(0.06)*q2
print(f'Левая граница доверительного интервала: {left}')

#Правая граница доверительного интервала
right = X1.mean() - X2.mean() - np.sqrt(0.06)*q1
print(f'Правая граница доверительного интервала: {right}')
```

```python
k = 0
for i in range(1000):
  X1 = sps.norm(loc=2, scale=1).rvs(size=25)
  X2 = sps.norm(loc=1, scale=np.sqrt(0.5)).rvs(size=25)
  left = X1.mean() - X2.mean() - np.sqrt(0.06)*q2
  right = X1.mean() - X2.mean() - np.sqrt(0.06)*q1
  if (left <= 1) and (right >= 1):
    k += 1

print(k)
```

# **Задание 2**

```python
#Выборка
X = sps.uniform(-5, 10).rvs(size=25)
print(f'Выборка: {X}')

#Первый член вариационного ряда
x1 = min(X)
print(f'Первый член вариационного ряда: {x1}')

#Квантили
q1 = sps.gamma(a=1).ppf([0.025])
q2 = sps.gamma(a=1).ppf([0.975])
print(f'Первый квантиль: {q1}')
print(f'Второй квантиль: {q2}')

#Левая граница доверительного интервала
left = x1 / ((2*q1 / 25) - 1)
print(f'Левая граница доверительного интервала: {left}')

#Правая граница доверительного интервала
right = x1 / ((2*q2 / 25) - 1)
print(f'Правая граница доверительного интервала: {right}')
```

```python
k = 0
for i in range(1000):
  X = sps.uniform(-5, 10).rvs(size=25)
  x1 = min(X)
  left = x1 / ((2*q1 / 25) - 1)
  right = x1 / ((2*q2 / 25) - 1)
  if (left <= 5) and (right >= 5):
    k += 1

print(k)
```
